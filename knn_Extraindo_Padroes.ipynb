As colunas CUST_ID e TENURE não trazem informações relevates. Por conta disso serão exluidas do DataFrame

```python
dados.drop(['CUST_ID','TENURE'], axis= 1, inplace= True)
```

```python
dados.head()
```

### **Tratando dados faltantes**

Podemos ver que a coluna MINIMUM_PAYMENTS possui vários dados faltantes. Para preencher estes dados, ferei o calculo da mediana, pegarei este valor e atribuirei a os dados que estão faltando na base de dados.



```python
mediana = dados['MINIMUM_PAYMENTS'].median()
dados.loc[dados['MINIMUM_PAYMENTS'].isnull(), 'MINIMUM_PAYMENTS'] = mediana
dados.isnull().sum()
```

Outra coluna que apresenta Dados Faltantes é a coluna CREDIT_LIMIT

Neste caso, farei o preenchimento com o Fillna tambem com os dados da mediana

```python
dados.fillna(dados['CREDIT_LIMIT'].median(), inplace= True)
dados.isnull().sum()
```

## Normalizando os Dados

```python
from sklearn.preprocessing import Normalizer
normalizar = Normalizer()

dados_normalizados = normalizar.fit_transform(dados)
```

```python
dados_normalizados
```

## Montando o Cluster

Aqui, apliquei o modelo e já salvei o grupo dentro de uma coluna no DataFrame original

Com isso, ganho tempo para fazer uma análise mais refinada ou mais aprofundada.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters= 10, n_init= 10, max_iter= 300)

dados['cluster'] = kmeans.fit_predict(dados_normalizados)
```

```python
dados
```

```python
dados.groupby(['cluster']).describe()
```

```python
centroides = kmeans.cluster_centers_

print(f'Quantidade de Centroides {len(centroides[0])}')
```

```python
max = len(centroides[0])
item = []
valor = []

for i in range(max):
    item.append(dados.columns.values[i])
    valor.append(centroides[:, i].var())

pd.options.display.float_format = '{:.2f}'.format

resultados = pd.DataFrame({'Item':item, 'Valor': valor})
resultados.sort_values(by= 'Valor', ascending= False)
```
